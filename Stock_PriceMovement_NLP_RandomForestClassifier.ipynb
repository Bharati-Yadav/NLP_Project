```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
from collections import Counter
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
from string import digits
```

```python
dataset=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
dataset.head(2)
```

```python
train = dataset[dataset['Date'] < '20150101']
test = dataset[dataset['Date'] > '20141231']
```

```python
data=train.iloc[:,2:27]
data.head(1)
```

```python
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
data.replace('[^\w\s]','')
```

```python
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(1)
```

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
newdata = pd.DataFrame(headlines,columns={'headlines'}) 
```

```python
newdata.head(5)
```

```python
newdata.keys()
```

```python
NSM = pd.concat([dataset.Label,newdata.headlines], axis=1)
```

```python
NSM.head()
```

```python
df_new = NSM[NSM['headlines'].notnull()]
```

```python
df_new['word_count']=df_new['headlines'].apply(lambda x:len(str(x).split(" ")))
df_new.head()
```

```python
df_new['char_count']=df_new['headlines'].str.len()
df_new[['headlines','char_count']].head()
```

```python
df_new['R']=df_new['headlines'].apply(lambda x:" ".join(x.lower() for x in x.split()))
df_new['R'].head()
```

```python
txt=df_new['headlines']
```

```python
def t(txt):
    string=" "
    for i in txt:
        string+=str(i)
    return string
t(txt[:1])
```

```python
remove_num=str.maketrans('','',digits)
new_txt=t(txt).translate(remove_num)
new_txt[:100]
```

```python
tokens=word_tokenize(new_txt)
tokens[:10]
```

```python
tokens_n = [word for word in tokens
             if word not in stopwords.words('english')]
tokens_n[:10]
```

```python
count_freq = Counter(tokens_n)
count_freq
```

```python
most_freq = count_freq.most_common(20)
most_freq
```

```python
string = ""
for word, freq in most_freq:
    string = string + " " + word
```

```python
wordcloud_est = WordCloud(width=1200, height=800,
                     background_color='black',
                     min_font_size=8)
wordcloud = wordcloud_est.generate(string)
```

```python
plt.figure(figsize=(12, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis=('off')
plt.tight_layout(pad = 0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix,accuracy_score
```

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(df_new.headlines)
```

```python
y=train.Label
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,y)
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
confusion_matrix(test['Label'],predictions)
```

```python
accuracy_score(test['Label'],predictions)
```

```python
precision_score(test['Label'], predictions)
```

```python
recall_score(test['Label'], predictions)
```

```python
f1_score(test['Label'], predictions)
```

```python
y_pred_prob = randomclassifier.predict_proba(test_dataset)[:, 1]
fpr, tpr, thresholds = roc_curve(test['Label'], y_pred_prob)
```

```python
plt.plot([0, 1], [1, 0], 'k--')
plt.plot(fpr, tpr, label='RandomForest Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.show()
```
